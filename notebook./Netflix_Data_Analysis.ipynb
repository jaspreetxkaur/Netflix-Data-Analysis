```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("netflix_titles.csv")

df.head()
```

```python
df.info() # checking the structure of the data
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=["country","date_added"])
```

```python
df["date_added"] = pd.to_datetime(df["date_added"], format='mixed')
```

```python
df["type"].value_counts().plot(kind="bar")

plt.title("Movies vs TV Shows on Netflix")
plt.xlabel("Type")
plt.ylabel("Count")

plt.show()
```

```python
top_countries = df["country"].value_counts().head(10)

top_countries.plot(kind="bar")

plt.title("Top Countries Producing Netflix Content")
plt.show()
```

```python
df["year_added"] = df["date_added"].dt.year

df["year_added"].value_counts().sort_index().plot()

plt.title("Netflix Content Growth Over Years")

plt.show()
```

```python
genres = df["listed_in"].str.split(",", expand=True).stack()

genres.value_counts().head(10).plot(kind="bar")

plt.title("Top Genres on Netflix")

plt.show()
```

```python
rating_counts = df['rating'].value_counts()

plt.figure(figsize=(8,5))
sns.barplot(x=rating_counts.index, y=rating_counts.values)

plt.xticks(rotation=45)
plt.title("Distribution of Netflix Content Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")

plt.show()
```

```python
content_year = df.groupby(['year_added','type']).size().unstack()

content_year.plot(figsize=(10,6))

plt.title("Movies vs TV Shows Added Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Titles")

plt.show()
```

```python
top_directors = df['director'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_directors.values, y=top_directors.index)

plt.title("Top Directors on Netflix")
plt.xlabel("Number of Titles")

plt.show()
```

```python
plt.figure(figsize=(10,6))

country_year = df.pivot_table(index='country', columns='year_added', aggfunc='size', fill_value=0)

sns.heatmap(country_year.head(10), cmap="coolwarm")

plt.title("Content Added by Country Over Years")

plt.show()
```

```python
#Key Insights

#1. Movies dominate the Netflix catalog.
#2. United States produces the largest number of titles.
#3. Content production increased dramatically after 2016.
#4. TV-MA rating is the most common.
#5. Drama and International movies are the most frequent genres.
```
